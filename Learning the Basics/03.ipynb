```python
import torch
```

### Zeros and ones

```python
# Creating a random tensor
random_tensor = torch.rand(size = (3,4))
print(random_tensor)
# Create a tensor of all zeros
zeros = torch.zeros(size=(3,4))
print(zeros)
```

```python
zeros * random_tensor
```

```python
# Creating a tensor of all ones
ones = torch.ones(size=(3,4))
print(ones)
print(ones.dtype)
print(ones*random_tensor)
```

### Creating a range of tensors and tensors-like

```python
# Use torch.range() and get a deprecated message, use torch.arange()
one_to_ten = torch.range(start=0,end=10, step=2)
print(one_to_ten)
torch.arange(0,100,5)
```

```python
torch.arange(0,10,2)
```

```python
# Creating tensors-like
ten_zeros = torch.zeros_like(input=one_to_ten)
ten_zeros
# It creates a tensor full of zeroes but as the same shape as the input
```

```python
ten_ones = torch.ones_like(input=one_to_ten)
ten_ones
```

### Summary:

```python
'''
Date - 14/Mar/2026
------------------------------------------------------------------------------------------------

          Attribute                                       Function
torch.range(start=,end=,step=)     Creates a tensor of the given range(start->end) with the  
                                                         given step
torch.zeros_like(input=)           Creates a tensor full of zeroes with the same shape as the
                                                        input tensor
torch.ones_like(input=)              Creates a tensor full of ones with the same shape as the
                                                        input tensor
'''
```

### Tensor Datatypes
**Note:** Tensor datatypes is one of the 3 big errors you'll run into with PyTorch & deep learning:

    1. Tensors not right datatype
    2. Tensors not right shape
    3. Tensors not on the right device

```python
# Float 32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0], 
                               dtype = None, # What datatype is the tensor (e.g. float32 or float 16)
                               device=None, # What device is your tensor on (e.g. CPU, GPU, or TPU)
                               requires_grad=False) # Whether or not to track gradients with these tensor operations.
float_32_tensor, float_32_tensor.dtype
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
int_32_tensor = torch.tensor([3,6,9], dtype=torch.int32)
float_32_tensor = torch.tensor([3,6,9] , dtype=torch.float32)
int_32_tensor * float_32_tensor
```

```python
int_32_tensor = torch.tensor([3,6,9], dtype=torch.long)
float_32_tensor = torch.tensor([3,6,9] , dtype=torch.float32)
int_32_tensor * float_32_tensor
```
